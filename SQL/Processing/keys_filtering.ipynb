```python
import pandas as pd
```

```python
#reading in csv file created containing all the country, ISO number and year columns from the original csv data sheets 
#downloaded from the web
spreadsheet1 = pd.read_csv('resources/keys_original.csv')
spreadsheet1
```

```python
#dropping duplicates
spreadsheet2 = spreadsheet1.drop_duplicates()
spreadsheet2
```

```python
#filtering to only include entries since 1961 (project will only include years since 1961)
spreadsheet3 = spreadsheet2.loc[(spreadsheet2['year'] >=1961)]
spreadsheet3
```

```python
#only keeping first country name listed for each instance of "iso" and "year"
spreadsheet4 = spreadsheet3.drop_duplicates(subset = ['iso', 'year'], keep='first')
spreadsheet4
```

```python
#creating the csv "keys" to be used in postres
spreadsheet4.to_csv('resources/keys.csv', index=False)
```
