```python
import pandas as pd
```

```python
#creating the dataframe
spreadsheet1 = pd.read_csv('Resources/GCB2022v27_MtCO2_flat.csv')
spreadsheet1
```

```python
#renaming columns to match postgres column names
spreadsheet1.columns = ["country", "country_code", "year", "total", "coal", "oil", "gas", "cement", "flaring", "other", "per_capita"]
spreadsheet1
```

```python
#filtering for dates from 1961
spreadsheet2 = spreadsheet1.loc[(spreadsheet1['year'] >=1961)]
spreadsheet2
```

```python
#filtering by country
spreadsheet3 = spreadsheet2[spreadsheet2['country_code'].isin(['AUS', 'AUT', 'BLR', 'BEL', 'BRA', 'BGR', 'CAN', 'CHN', 'HRV', 
    'CYP', 'CZE', 'DNK', 'EST', 'FIN', 'FRA', 'DEU', 'GRC', 'HUN','ISL', 'IRL', 'ITA', 'JPN', 'KAZ', 'LVA', 'LIE', 'LTU', 'LUX',
       'MLT', 'NLD', 'NZL', 'NOR', 'POL', 'PRT', 'KOR', 'ROU', 'RUS',
       'SVK', 'SVN', 'ESP', 'SWE', 'CHE', 'TUR', 'UKR', 'GBR',
       'USA', 'WLD'])]
spreadsheet3
```

```python
spreadsheet3.to_csv('resources/emissions_since1961.csv', index=False)
```
