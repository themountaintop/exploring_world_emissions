```python
# Let's try out OneHotEncoder!!!!!!!
from sklearn.preprocessing import OneHotEncoder

#instance of OneHotEncoder
enc = OneHotEncoder()

# fit and transform 'country_code' column
encoded = enc.fit_transform(df[['country_code']])

# convert the encoded data to a DataFrame
encoded_df = pd.DataFrame(encoded.toarray(), columns=enc.get_feature_names_out(['country_code']))

# concatenating the encoded_df with the OG df:
copy_df = pd.concat([df, encoded_df], axis=1)

# drop the original 'country_code' column
copy_df.drop(['country_code'], axis=1, inplace=True)
copy_df.head(55)
```

```python
# Filling NA values == 0:
copy_df = copy_df.fillna(value=0)
copy_df.head()

# Not sure why I had to do this again.
```

```python
from sklearn.model_selection import train_test_split

# filtering df for less than or equal to 2011:
df_train = copy_df.query("year <= 2011")

# filtering df for greater than 2011:
df_test = copy_df.query("year > 2011")

# Splitting for feature and target variables:
X_train = df_train.drop(['temp_change'], axis=1)
y_train = df_train['temp_change']

X_test = df_test.drop(['temp_change'], axis=1)
y_test = df_test['temp_change']
```

```python
# Linear Regression model:
from sklearn.linear_model import LinearRegression

# Instance of LinearRegression
model = LinearRegression()

# fit the model w/ training data
model.fit(X_train, y_train)

# predict on test data
y_pred = model.predict(X_test)
```

```python
# Evaluating model (MSE):
from sklearn.metrics import mean_squared_error

# calculate MSE:
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
# Evaluating model (r-Squared):
from sklearn.metrics import r2_score

# calculate r-squared score
r_squared = r2_score(y_test, y_pred)
print(f"R-Squared Score: {r_squared}")
```

```python
# Evaluating model (RMSE):
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```
