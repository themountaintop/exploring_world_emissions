```python
import pandas as pd
```

```python
# Load data
temp_change_df = pd.read_csv("../Resources/Annual_Surface_Temperature_Change.csv")
```

```python
temp_change_df.head()
```

```python
temp_change_df.info()
```

```python
# Let's take a look at what values are in some of these columns
for col in ['Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor']:
    print(temp_change_df[col].value_counts())
    print("--------------------")
```

```python
# Since these columns all have the same values, we can pull them out as a data series
temp_change_metadata = temp_change_df[['Indicator', 'Unit', 'Source', 'CTS_Code', 
                                       'CTS_Name', 'CTS_Full_Descriptor']].iloc[0]
temp_change_metadata
```

```python
# Now we can get rid of these rows as well as the ISO2 and ObjectID
temp_change_years = temp_change_df.drop(columns=['ObjectId', 'ISO2', 'Indicator', 'Unit', 'Source', 
                                         'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor'])
temp_change_years.head()
```

```python
# Rename the year columns to be just the number
for col in range(2, len(temp_change_years.columns)):
    temp_change_years.rename(columns={temp_change_years.columns[col]:temp_change_years.columns[col][1:]}, inplace=True)
```

```python
temp_change_years.columns
```

```python
# Transfer the year columns to rows
temp_change = temp_change_years.melt(id_vars=['Country', 'ISO3'], var_name='Year', value_name='Temp_Change')
```

```python
temp_change.info()
```

```python
# Change year to int datatype 
temp_change['Year'] = temp_change['Year'].astype('int')
```

```python
temp_change.info()
```

```python
temp_change
```

```python
# Export data and metadata
temp_change.to_csv("../Resources/Temperature_Change.csv")
temp_change_metadata.to_csv("../Resources/Temperature_Change_Metadata.csv")
```
