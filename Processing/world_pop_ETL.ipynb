```python
# Dependencies:
import pandas as pd
```

```python
# Load CSV and sample:
emission_df = pd.read_csv("../Resources/GCB2022v27_MtCO2_flat.csv")
# emission_df[(emission_df['Year'].values >= 1965) & (emission_df['Country'].values == 'USA')]
emission_df
```

```python
# Check null and drop any null values:
emission_df.isnull().sum()
```

```python
emission_df = emission_df.dropna(how='any')
emission_df
```

```python
# Rename column ISO 3166-1 alpha-3 to Country Code:
emission_df = emission_df.rename(columns={'ISO 3166-1 alpha-3':'Country Code', 'Per Capita': 'PerCapita'})
emission_df.sample(25)
```

```python
emission_df.to_csv('emission_cleaned')
```

```python
# Load CSV and drop columns not needed:
world_pop_df = pd.read_csv("../Resources/world_pop.csv")
world_pop_df = world_pop_df.drop(columns=["Indicator Code","Indicator Name"])
world_pop_df
```

```python
# Stacked
world_pop_df3 = pd.DataFrame(world_pop_df.set_index(['Country Name', 
                                                     'Country Code']).stack())
world_pop_df3.sample(100)
```

```python
world_pop_df3.to_csv('world_population_by_year')
```
